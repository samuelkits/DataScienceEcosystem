Exploratory Data Analysis (EDA) is a crucial step in the data analysis process that involves summarizing the main characteristics of a dataset, often visualizing them to uncover patterns, spot outliers,  with the help of summary statistics and graphical representations

```python
import pandas as pd
import numpy as np
```

```python
#Read the file
df=pd.read_csv('EDA.csv',na_values=["??","????"],index_col=[0])
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
#Remove irrelevant column 'Automatic'(inplace = True)
df.drop('Automatic',axis = 1,inplace = True)
print(df)
```

```python
df.shape
```

```python
#Remove duplicate rows and analyze
df.drop_duplicates(keep ='last',inplace=True)
df.shape
```

```python
#Rename colum price to selling price
df.rename(columns={'Price':'Sellingprice'},inplace = True)
df.head()
```

```python
#Drop missing values of row and observe
df.dropna(axis=0,inplace = True)
df.shape
```

```python
#Calculate measure of central tendency - Age (Mean,Median,Mode)
print('Mean of Age is:',df['Age'].mean())
print('Median of Age is:',df['Age'].median())
print('Mode of Age is:',df['Age'].mode())
```

```python
#Calculate variability(Spread/Distribution) measures for Age - range, variance, standard deviation
m = df['Age'].max()
n = df['Age'].min()
print(m,n)
print('Range of Age is',m-n)
print('Variance of Age is:',df['Age'].var())
print('Standard deviation of Age is:',df['Age'].std())
```

```python
#Calculate IQR(Inter Quartile Range) using quantile for Age[Q3-Q1]
Q1 = df['Age'].quantile(0.25)
Q2 = df['Age'].quantile(0.50)
Q3 = df['Age'].quantile(0.75)
IQR = Q3-Q1
print(Q2)
print('IQR of Age is:',IQR)
```

```python
#Calculate z-score for Age[Z = (X-mean)/sd]
import scipy as sp
df1 = df.copy()
df1['Zscore'] = sp.stats.zscore(df1['Age'])

df1.tail()
```

```python
#Plot heatmap using the correlation
import seaborn as sns
import numpy as np
df_numeric = df.select_dtypes(include=[np.number])
c = df_numeric.corr()
sns.heatmap(c,xticklabels= c.columns,yticklabels=c.columns,annot=True)
```

```python
#correlation for subset of columns
t = df[['Age','Sellingprice','KM']]
c = t.corr()
sns.heatmap(c,xticklabels= c.columns,yticklabels=c.columns,annot=True)
```

```python
#Add two rows at the end of the dataframe
#append a row with new value
df2 = {'Sellingprice':12400, 'Age':90.0, 'KM':45362.00, 'FuelType':'Diesel', 'HP':90.0, 'MetColor':1.0, 'Automatic':1, 'CC':1800,
       'Doors':'3', 'Weight':1165}
df3 = {'Sellingprice':14000, 'Age':100,'KM':45362.00, 'FuelType':'Diesel', 'HP':90.0, 'MetColor':1.0, 'Automatic':1, 'CC':1800,
       'Doors':'3', 'Weight':1165}
# Convert the dictionary to a DataFrame
df2 = pd.DataFrame([df2])
df3 = pd.DataFrame([df3])
# Concatenate the new row to the copied DataFrame
df = pd.concat([df, df2], ignore_index=True)
df = pd.concat([df, df3], ignore_index=True)
df.tail()
```

```python
df['Age'].fillna(0,inplace = True)
df.tail()
```

```python
#Detect outliers in Age column
sns.catplot(x='Age',kind='box',data=df)
print('Mean',df['Age'].mean())
print('Median',df['Age'].median())
print('Mode',df['Age'].mode())
```

```python
#Remove the outliers by recalculating IQR in updated 'Age' and analyze with boxplot
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3-Q1
print('IQR of Age is:',IQR)
l = Q1-1.5*IQR #lower boundary outlier
u = Q3+1.5*IQR
print('Lower limit is:',l)
print('Upper limit is:',u)
df_out = df[(df['Age'] > l) & (df['Age'] < u)]

sns.catplot(x='Age',kind='box',data=df_out)
```

```python
#Remove outliers using z-score by recalculating z-score in Age column
zscore = sp.stats.zscore(df['Age'])
print(zscore)

filtered = (zscore < 3 and zscore >-3)
new_df = df[filtered]

sns.catplot(x='Age',kind='box',data=new_df)
```
